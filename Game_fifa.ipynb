```
import pyspark.pandas as ps
#csv
df = ps.read_csv(local_arquivo)

df.head()
```

#DROP DUPLICATES AND ORDERING

```
df_duplicated = df.duplicated()

df = df.drop_duplicates()
df = df.sort_values('Overall', ascending=False)
```

# SELECTING OVERALL AVERAGE BY AGE

```
df_age_overall = ps.sql('''
    SELECT AVG(Overall) AS Mean_overall, Age, count(Age) AS Number_players from {DF} GROUP BY Age HAVING COUNT(Age) > 35 ORDER BY Mean_overall DESC

''', DF=df)

df_age_overall.head(20)
```

```
df_age_overall.plot.scatter(x='Age', y='Mean_overall', trendline='ols', trendline_color_override='red')
```

# SELECTING OVERALL AVERAGE BY COUNTRY (ONLY COUNTRY WITH MORE THAN 50 PLAYERS)

```
df_country_groupby_overall = ps.sql('''
    SELECT Nationality, AVG(Overall) AS Mean_overall from {DF} group by Nationality HAVING COUNT(Nationality) > 50 order by Mean_overall desc
''', DF=df)

df_country_groupby_overall.head(18)
```

# CHANGING COLUMN NAMES

```
columns_before = df.columns
columns_after = df.columns.str.replace(' ', '_')
for column in df.columns:
    columns_after = df.columns.str.replace(' ', '_')
    df = df.rename(columns={column: column.replace(" ", "_")})

```

```
df.head()
```

# SELECTING OVERALL AVERAGE BY CLUB NAME

```
df_avg_team = ps.sql('''
    SELECT AVG(Overall) AS Avg, Club_Name FROM {DF} group by Club_name order by Avg DESC
''',DF=df)

df_avg_team.head(15)
```

# FUNCTION TO SELECT PLAYER FROM ANY TEAM

```
def all_team(team):
    return ps.sql('''
    SELECT * FROM {DF} WHERE Club_Name = {team}
    ''',DF=df, team=team)

all_team('Arsenal').head()
```
